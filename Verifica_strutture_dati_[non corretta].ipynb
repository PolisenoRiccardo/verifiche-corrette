##Esercizio 1

La software house PuntoDoc sta seguendo tre progetti in cui sono impegnate diverse persone. il progetto IoT vede impegnati Aldo, Marco, Stefano e Maria. Il progetto Cloud vede impegnati Stefania, Carlo, Maria e Andrea. Il progetto Database vede impegnati Carlo, Maria, Antonio, Stefano, Stefania e Davide. Due dipendenti, Ornella e Giulia, attualmente non sono impegnati in nessun progetto.

1. memorizzare le informazioni del testo precedente in una opportuna struttura dati, tenendo anche conto delle richieste degli esercizi successivi

```python
IoT = {'Aldo', 'Marco', 'Stefano', 'Maria'}
cloud = {'Andrea', 'Carlo', 'Stefania', 'Maria'}
database = {'Carlo', 'Maria', 'Antonio', 'Stefano', 'Stefania', 'Davide'}
```

2. quanti sono i dipendenti della società PuntoDoc?

```python
PuntoDoc = database.union(IoT, cloud)
len(PuntoDoc)
```

3. la dirigenza vuole premiare i dipendenti che lavorano in tutti i progetti. Chi sono?

```python
print(PuntoDoc.intersection(IoT, cloud, database))
```

4. quali sono i dipendenti che lavorano ad un solo progetto?

```python
unProgetto = IoT.difference(cloud ,database)
unProgetto.update(cloud.difference(IoT ,database))
unProgetto.update(database.difference(IoT ,cloud))
print(unProgetto)
```

5. quali sono i dipendenti che lavorano al progetto IoT ma non lavorano al progetto Database?

```python
print(IoT.difference(database))
```

##Esercizio 4

1. caricare in una lista 100 lettere casuali appartenenti all'alfabeto inglese

```python
import random

rndlettere = [chr(random.randint(ord('a'), ord('z'))) for i in range(100)]
print(rndlettere)
len(rndlettere)
```

2. eliminare i duplicati

```python
rndlettere = list(set(rndlettere))
len(rndlettere)
```

3. eliminare le vocali e visualizzare le lettere rimaste

```python
for lettera in rndlettere:
  if lettera == 'a' or lettera == 'e' or lettera == 'i' or lettera == 'o' or lettera == 'u':
    rndlettere.remove(lettera)
print(rndlettere)
len(rndlettere)
```
